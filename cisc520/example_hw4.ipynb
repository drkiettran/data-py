# Iris dataset from HW4
This dataset has no `label` or `y` named `species`. Unsupervised learning model is to predict this value from the dataset.

```python
filename = '1_Iris_Test_data.txt'
pd_test = pd.read_csv(filename, header=None, sep=" ", names=["sepal.length", "sepal.width", "petal.length", "petal.width"])

```

```python

k_means = cluster.KMeans(n_clusters=3, max_iter=50, random_state=1)
k_means.fit(pd_test) 
labels = k_means.labels_
pd_out = pd.DataFrame(labels, columns=['Cluster ID'])
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(pd_out)
```

# Actual Iris dataset from UCI
This dataset has `species` as a `label` or `y`. `Species` column is dropped before applying K-Means algorithm. This is for comparison purposes. Inspect the actual dataset and the `predicted` values to see there are some incorrect prediction. They would be the errors of using K-Means

```python
filename = 'iris.data'
pd_iris = pd.read_csv(filename, header=None, sep=",", names=["sepal.length", "sepal.width", "petal.length", "petal.width", "species"])
pd_iris
```

```python
data = pd_iris.drop('species', axis=1).dropna() # Drop both Species column and NaN
data
```

```python
# k_means = cluster.KMeans(n_clusters=3, max_iter=50, random_state=1)
k_means = cluster.KMeans(n_clusters=3, random_state=0)
k_means.fit(data) 
labels = k_means.labels_
data_out = pd.DataFrame(labels, columns=['Cluster ID'])
# with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#     print(data_out)
# Make the output from 0, 1, 2 to 1, 2, 3.
data_array = data_out.to_numpy()
data_array = data_array + 1
data_array
```

```python
# Experimenting with just using sepal.length and sepal.width.
sepal_data = pd_iris.drop(['petal.length', 'petal.width', 'species'], axis=1)
sepal_data
```

```python
k_means = cluster.KMeans(n_clusters=3, random_state=0)
k_means.fit(sepal_data) 
labels = k_means.labels_
data_out = pd.DataFrame(labels, columns=['Cluster ID'])

# Make the output from 0, 1, 2 to 1, 2, 3.
data_array = data_out.to_numpy()
data_array = data_array + 1
data_array
```

```python
petal_data = pd_iris.drop(['sepal.length', 'sepal.width', 'species'], axis=1)
petal_data
```

```python
k_means = cluster.KMeans(n_clusters=3, random_state=0)
k_means.fit(petal_data) 
labels = k_means.labels_
data_out = pd.DataFrame(labels, columns=['Cluster ID'])

# Make the output from 0, 1, 2 to 1, 2, 3.
data_array = data_out.to_numpy()
data_array = data_array + 1
data_array
```

# Image data

```python
filename = '2_Image_Test.txt'
pd_image = pd.read_csv(filename, header=None, sep=",")
pd_image = pd_image.dropna() # Removing NaN
pd_image
```
